Functions

```python
def corr2D(tas,hurs):
    corrMatrix=np.zeros((90,134))
    for lat in range(90):
        for lon in range(134):
            corrMatrix[lat,lon]=np.corrcoef(tas['tas'][:,lat,lon],hurs['hurs'][:,lat,lon])[1,0]
    return corrMatrix

def load(base_path,gcm,rcp,rp,version,date):
    tas_path = base_path + "/" + gcm + "/tas/" + rcp + "/" + "remapped_tas_patterns_EUR-11_"+ gcm + "_"+ rcp+ "_"+ rp+ "_SMHI-RCA4_"+ version+ "_day_"+ date + ".nc"
    hurs_path = base_path + "/" + gcm + "/hurs/" + rcp + "/" + "remapped_hurs_patterns_EUR-11_" + gcm + "_" + rcp+ "_"+ rp + "_SMHI-RCA4_"+ version+ "_day_"+ date + ".nc"
    try:
        tas = xr.open_dataset(tas_path)
        #tas = tas.to_dataframe()
        try:
            hurs = xr.open_dataset(hurs_path)
            #hurs = hurs.to_dataframe()
        except:
            print("Can't load hurs: " + hurs_path)
            hurs=[]
    except:
        print("Can't load tmp: "+ tas_path)
        tas=[]

    #print(list(tas['time_bnds']))
    return tas, hurs
```

```python
gcm="ICHEC-EC-EARTH"
rcp="rcp45"
rp="r12i1p1"
version="v1"
date="20060101-20101231"

tas_data,hurs_data=load(base_path,gcm,rcp,rp,version,date) # Load the tas and hurs data
corrMatrix=corr2D(tas_data,hurs_data)
```
